```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split    
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import joblib
```

```python
df_incoming_bikes = pd.read_pickle('incoming_bikes.pkl')
df_outgoing_bikes = pd.read_pickle('outgoing_bikes.pkl')
```

```python
df_incoming_bikes.head(100)
```

```python
df_outgoing_bikes.head(100)
```

```python
df_outgoing_bikes.head(100)
min_value = df_outgoing_bikes['COUNTEVENTS'].min()
max_value = df_outgoing_bikes['COUNTEVENTS'].max()

df_outgoing_bikes['COUNTEVENTSNORM'] = (df_outgoing_bikes['COUNTEVENTS'] - min_value) / (max_value - min_value)
```

```python
# outputs.head(100)
```

```python
#plot start hour x axis, count events y axis
# plt.scatter(inputs['STARTHOUR'], outputs, color='red', marker='+')
df_test = pd.DataFrame(df_outgoing_bikes.groupby(['STARTSTATIONID', 'STARTMONTHNUM', 'STARTHOUR', 'DAYOFWEEK'], as_index=False)['COUNTEVENTS'].mean())
#normalize count events
min_value = df_test['COUNTEVENTS'].min()
max_value = df_test['COUNTEVENTS'].max()

df_test['COUNTEVENTSNORM'] = (df_test['COUNTEVENTS'] - min_value) / (max_value - min_value)



# plt.scatter(df_test['STARTMONTHNUM'], df_test['COUNTEVENTS'], color='blue', marker='+')
inputs = df_test.drop(['COUNTEVENTS', 'COUNTEVENTSNORM'], axis='columns')
outputs = df_test['COUNTEVENTSNORM']

inputs_train, inputs_test, outputs_train, outputs_test = train_test_split(inputs, outputs, test_size=0.2)
#same plot but average 
print(df_test.shape)
df_test.head(100)
```

```python
# model_outgoing_bikes = LinearRegression()
model_outgoing_bikes = RandomForestRegressor(n_estimators=100, random_state=42, verbose=0)
# model_outgoing_bikes = DecisionTreeClassifier()
model_outgoing_bikes.fit(inputs_train, outputs_train)
score = model_outgoing_bikes.score(inputs_test, outputs_test)
print(score)
```

```python
#For incoming bikes
df_test = pd.DataFrame(df_incoming_bikes.groupby(['ENDSTATIONID', 'ENDMONTHNUM', 'ENDHOUR', 'DAYOFWEEK'], as_index=False)['COUNTEVENTS'].mean())
min_value = df_test['COUNTEVENTS'].min()
max_value = df_test['COUNTEVENTS'].max()
df_test['COUNTEVENTSNORM'] = (df_test['COUNTEVENTS'] - min_value) / (max_value - min_value)



# plt.scatter(df_test['STARTMONTHNUM'], df_test['COUNTEVENTS'], color='blue', marker='+')
inputs = df_test.drop(['COUNTEVENTS', 'COUNTEVENTSNORM'], axis='columns')
outputs = df_test['COUNTEVENTSNORM']

inputs_train, inputs_test, outputs_train, outputs_test = train_test_split(inputs, outputs, test_size=0.2)
#same plot but average 
print(df_test.shape)
df_test.head(100)
```

```python
# model_outgoing_bikes = LinearRegression()
model_incoming_bikes = RandomForestRegressor(n_estimators=100, random_state=42, verbose=0)
# model_outgoing_bikes = DecisionTreeClassifier()
model_incoming_bikes.fit(inputs_train, outputs_train)
score = model_incoming_bikes.score(inputs_test, outputs_test)
print(score)
```
